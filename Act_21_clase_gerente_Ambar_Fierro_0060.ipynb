```python
# 1. Definimos el molde (la Clase)
class Gerente:
    def __init__(self, nombre, liderazgo, vision, adaptabilidad):
        # Estos son los datos que guardamos de cada gerente (Atributos)
        self.nombre = nombre
        self.liderazgo = liderazgo
        self.vision = vision
        self.adaptabilidad = adaptabilidad

    # 2. Definimos las acciones (Funciones)
    def planificar(self):
        print(f"{self.nombre} está creando el plan para el próximo mes.")

    def gestionar_talento(self):
        print(f"{self.nombre} está ayudando a los empleados a mejorar.")

    def evaluar_resultados(self):
        print(f"{self.nombre} está revisando si se cumplieron las metas.")

# 3. Creamos a las personas reales (Objetos)

# Creamos a la primera gerente con sus datos
gerente1 = Gerente("Ana López", "Muy humana", "Ver el futuro del negocio", "Se ajusta rápido")

# Creamos al segundo gerente con sus datos
gerente2 = Gerente("Luis Pérez", "Directo", "Ahorrar dinero", "Mantiene la calma")

# 4. Ponemos a los gerentes a trabajar (Mensajes)


print("--- Turno de la Gerente 1 ---")
print(f"Ella es {gerente1.nombre} y su visión es: {gerente1.vision}")
gerente1.planificar() # Llamamos a la función

print("\n--- Turno del Gerente 2 ---")
print(f"Él es {gerente2.nombre} y su liderazgo es: {gerente2.liderazgo}")
gerente2.evaluar_resultados() # Llamamos a la función

print("programa terminado")
print(" clase gerente por Ambar Fierro NL = 19")
```
