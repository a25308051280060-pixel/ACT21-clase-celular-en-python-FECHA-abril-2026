```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")

# Instanciación
celular1 = Celular("Samsung", "Galaxy S23", 800)
celular2 = Celular("apple", "i phone 15", 1000)

#llamadas a las funciones para cada objeto celular 1 y celular 2
print("celular1:")
celular1.mostrar_info()
celular1.encender()

print("celular2:")
celular2.mostrar_info()
celular2.encender()
print("creador de software por Ambar Fierro NL = 19")
print(" programa terminado")
```
